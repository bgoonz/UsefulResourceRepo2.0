## Tensors

`tf.Tensor` has a `shape` and a `dtype`. The `device` displays where the Tensor is hosted in memory. You can manually 

```python
x = tf.random.uniform((3,))
print(x.numpy())
print(x.shape, x.dtype)
print(x.device)

with tf.device('CPU:0'):
    y = tf.random.uniform((3,))
print(y.device)
```
