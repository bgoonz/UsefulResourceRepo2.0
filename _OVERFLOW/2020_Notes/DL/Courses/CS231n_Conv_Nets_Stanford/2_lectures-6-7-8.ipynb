### Activation Functions

#### Sigmoid
$$\sigma(x) = \frac{1}{(1+e^{-x})}$$

- Squashes numbers to range [0,1]
- Has a *"linear regime"* near zero that looks like a linear function

Problems:
- Easy to saturate (kills the gradient)

__How Sigmoid Neurons Saturate__

$$x = 10$$

$$\sigma(x) = 0.000045$$


$$\frac{\partial \sigma}{\partial x} = \sigma(1-\sigma) = 0.000045$$

$$\frac{\partial L}{\partial x} = \frac{\partial L}{\partial \sigma}\frac{\partial \sigma}{\partial x} = \frac{\partial L}{\partial \sigma}*0.000045\approx 0$$

```python
z = 1/(1+np.exp(10))
print(z)
print(z*(1-z))
```

Now the weights $W$ in this neuron will only be updated a very small amount... ensuring the activation doesnt change... meaning the weights can never update again. __All prior nodes get propagated a near-zero gradient__

Very positive or negative activations can saturate the sigmoid-activated neuron. Likewise, large update steps can jump parameters into saturation.

__Sigmoid outputs are  not zero-centered__

What happens when inputs to a neuron are always positive. $\rightarrow$ gradients will either be all positive or all negative.

$$f(W\cdot X + b)$$

$$\frac{\partial Z}{\partial W} = X \quad(X all positive)$$  

$$\frac{\partial L}{\partial W} = \frac{\partial L}{\partial W} \cdot X$$

So the gradient of W will always be the sign of the upstream gradient coming down. Either increase all values of W by a positive amount or decrease all by neg amount. Leads to a zig-zagging pattern of optimization that is ineffcient.

Want input X __to be zero-centered__ so that gradients can take both positive and negative values. __Want activations to be both positive and negative__.

#### tanh

- Squashes numbers to range [-1,1]
- flat regions in activation functionstill lead to dead neurons
- zero-centered


#### ReLU

$$f(x) = max(0,x)$$

- Does not saturate in the positive region
- computationally efficient
- converges faster than sigmoid and tanh

Started to be used a lot around 2012.

- __not zero centered__

The gradient when X<0 $\rightarrow$ ZERO

This means that these neurons will neither update nor activate. Large updates can lead to neurons being nocked off the data manifold.

#### Leaky ReLU

$$f(x) = max(0.01x,x)$$

- all the benefits of ReLU
- will not die

#### Paremetric ReLU (PRELU)

$$f(x) = max(\alpha x ,x)$$

#### Exponential Linear Units

$$f(x) = x \, or \, \alpha(exp(x)-1)$$

#### Maxout Neuron

Generalizes ReLU and Leaky ReLU

$$f(x) = max(w_1x+b_1, w_2x+b_2)$$

```python
class Neuron(tf.layers.Layer):
    """A single Neuron: Dense layer with one hidden unit"""
    def __init__(self, input_shape):
        super(Neuron, self).__init__()
        self.w = tf.Variable(initial_value=tf.random_normal_initializer()(shape=(input_shape,1),
                                                                                dtype=tf.float32),
                                                                          trainable=True)
        self.b = tf.Variable(initial_value-tf.zero_initializer()(shape=(1,),
                                                                dtype=tf.float32),
                                                                 trainable=True)
        def call(self, X):
            return tf.matmul(X, self.w) + self.b
```

## Commonly Used Activation Functions

### Sigmoid
$$\sigma(x) = \frac{1}{1+e^{-x}}$$

Squashes input to the range of [0,1]. Two major issues. 

__Sigmoid layer kills gradients__ - When the neuron activation saturates at 0 or 1 (Linear activation very big or very small) then the gradient is almost zero. (see below) 

We know that backprop through a sigmoid activation results in the gradient will result in a multiplication of the very small gradient $\frac{\delta \sigma}{\delta x}$, as a result, all prior layers receive a zeroed gradient and fail to update any further. 

$$\frac{\delta e}{\delta x} = \frac{\delta \sigma}{\delta x}\frac{\delta e}{\delta y}$$

Also, when weights are initialized in layers ending with sigmoid activation, if weights are initialized to high/low, the neuron can be saturated at the very start of training.

__Sigmoid outputs are not zero centered__ Later layers in the network will receive inputs is always positive. Because of this later gradients will either be all positive or all negative



```python
def plot_function(func, func_name):
    x = tf.range(-5.,5., delta=.25)
    with tf.GradientTape() as t:
        t.watch(x)
        y = func(x)

    gradient = t.gradient(y, x) 

    plt.title("{} Activation".format(func_name))
    plt.plot( x.numpy(), y.numpy())
    plt.ylabel("{}(x)".format(func_name))
    plt.xlabel("X")
    plt.show()

    plt.title("{} Gradient".format(func_name))
    plt.plot( x.numpy(), gradient.numpy())
    plt.ylabel("delta_{}/delta_x".format(func_name))
    plt.xlabel("X")
    plt.show()
```

```python
#tf.nn.sigmoid()
def sigmoid(x):
    return tf.divide(1,tf.add(tf.exp(-x),1))

plot_function(sigmoid, "Sigmoid")
```

### Tanh

Squashes the input to the range [-1,1] Its activations also saturate, but its output is zero centered.

$$tanh(x) =\frac{e^x-e^{-x}}{e^x+e^{-x}}$$
$$= 2\sigma (2x) - 1$$

```python
#tf.nn.tanh()
def tanh(x):
    return tf.add(tf.multiply(2,sigmoid(tf.multiply(2,x))),-1)

plot_function(tanh, "tanh")
```

### ReLU

Max is a cheap operation compared to exponential. __ReLU units can die__ a large gradient through a ReLU might cause the weights to update such that these neurons will never activate on a data point again.


$$ReLU(x)=max(0,x)$$

```python
#tf.nn.relu()
def ReLU(x):
    return tf.math.maximum(0,x)

plot_function(ReLU, "ReLU")
```

### Leaky ReLU

Applies a small negative slope for negative input valies. This paralmeter $\alpha$ can be a learnable parameter as in PRELU activation. Research is not clear whether this is helpful to training.

$$Leaky ReLU(x)=max(\alpha x, x)$$

```python
#tf.nn.leaky_relu()
def Leaky_ReLU(x, alpha=0.05):
    return  -tf.multiply(alpha, ReLU(-x)) + ReLU(x)

plot_function(Leaky_ReLU, "Leaky_ReLU")
```
